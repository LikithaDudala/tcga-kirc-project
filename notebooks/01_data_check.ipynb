# Step 2: Identify Patient ID Columns
We need to find columns that contain patient identifiers so we can later link clinical, follow-up, and expression data.

```python
# Find columns in clinical data that look like patient IDs
# We search for columns containing "id", "case", "patient", or "submitter"
id_keywords = ["id", "case", "patient", "submitter", "barcode"]

print("=== Clinical columns that may contain patient IDs ===")
for col in clinical.columns:
    if any(kw in col.lower() for kw in id_keywords):
        print(f"  {col}")

print()
print("=== Follow-Up columns that may contain patient IDs ===")
for col in follow_up.columns:
    if any(kw in col.lower() for kw in id_keywords):
        print(f"  {col}")
```

```python
# Show 5 example values from the likely ID columns in clinical data
# "case_submitter_id" is the typical TCGA patient ID column (format: TCGA-XX-XXXX)
print("=== Clinical — sample ID values (first 5) ===")
for col in clinical.columns:
    if any(kw in col.lower() for kw in ["submitter_id", "case_id", "patient"]):
        print(f"\n  {col}:")
        print(f"  {clinical[col].head().tolist()}")
```

```python
# Show 5 example values from the likely ID columns in follow-up data
print("=== Follow-Up — sample ID values (first 5) ===")
for col in follow_up.columns:
    if any(kw in col.lower() for kw in ["submitter_id", "case_id", "patient"]):
        print(f"\n  {col}:")
        print(f"  {follow_up[col].head().tolist()}")
```

```python
# Inspect expression data column names
# Expression columns are usually sample barcodes like TCGA-XX-XXXX-01
# The first column might be a gene name/ID column

print("=== Expression Data — first 5 column names ===")
print(expression.columns[:5].tolist())

print("\n=== Expression Data — last 5 column names ===")
print(expression.columns[-5:].tolist())

print(f"\nTotal columns in expression data: {len(expression.columns)}")
```

```python
# Extract patient IDs from expression column names
# TCGA barcodes look like: TCGA-XX-XXXX-01A
# The patient ID is the first 3 parts: TCGA-XX-XXXX
# We remove the sample portion (-01, -11, etc.) by keeping only the first 12 characters
# or by splitting on "-" and joining the first 3 parts

# Get columns that look like TCGA barcodes (skip the gene name column)
tcga_columns = [col for col in expression.columns if col.startswith("TCGA")]

print(f"Number of TCGA sample columns: {len(tcga_columns)}")
print(f"\nExample full barcodes (first 5):")
print(tcga_columns[:5])

# Extract patient ID by keeping first 3 parts: TCGA-XX-XXXX
expression_patient_ids = ["-".join(col.split("-")[:3]) for col in tcga_columns]

print(f"\nExtracted patient IDs (first 5):")
print(expression_patient_ids[:5])
```

```python
# Count unique patients in each dataset
# This helps us understand how many patients overlap across datasets

# Clinical: use case_submitter_id (adjust if your column name is different)
id_col_clinical = [c for c in clinical.columns if "submitter_id" in c.lower()][0]
clinical_ids = clinical[id_col_clinical].nunique()

# Follow-up: use the submitter_id column
id_col_followup = [c for c in follow_up.columns if "submitter_id" in c.lower()][0]
followup_ids = follow_up[id_col_followup].nunique()

# Expression: unique patient IDs we extracted above
expr_ids = len(set(expression_patient_ids))

print("=== Unique Patient Counts ===")
print(f"  Clinical ({id_col_clinical}):    {clinical_ids}")
print(f"  Follow-Up ({id_col_followup}):   {followup_ids}")
print(f"  Expression (from barcodes):       {expr_ids}")
```

# Step 3: Filter Expression Data — Keep Tumor Samples Only
In TCGA, sample barcodes end with a sample type code:
- **-01** = Primary Tumor
- **-11** = Solid Tissue Normal

For survival analysis we only need tumor samples, so we drop the normal (-11) columns.

```python
# Count how many columns are tumor (-01) vs normal (-11) vs other
# The last 3 characters of a TCGA barcode column indicate sample type

tumor_cols  = [col for col in tcga_columns if col.endswith("-01")]
normal_cols = [col for col in tcga_columns if col.endswith("-11")]
other_cols  = [col for col in tcga_columns if not col.endswith("-01") and not col.endswith("-11")]

print(f"Tumor samples  (-01): {len(tumor_cols)}")
print(f"Normal samples (-11): {len(normal_cols)}")
print(f"Other samples:        {len(other_cols)}")

# Show a few examples of each type
print(f"\nExample tumor columns:  {tumor_cols[:3]}")
print(f"Example normal columns: {normal_cols[:3]}")
```

```python
# Build the filtered expression dataframe
# WHY tumor only? In survival analysis we want to link each patient's
# tumor gene expression to their survival outcome. Normal tissue samples
# don't reflect the tumor biology driving patient outcomes.

# Step 1: Identify the gene name column (the first non-TCGA column)
gene_col = [col for col in expression.columns if not col.startswith("TCGA")]
print(f"Gene name column(s): {gene_col}")

# Step 2: Keep gene column + tumor sample columns only
cols_to_keep = gene_col + tumor_cols
expression_tumor = expression[cols_to_keep]

print(f"\nOriginal expression shape:  {expression.shape}")
print(f"Filtered expression shape:  {expression_tumor.shape}")
print(f"  (rows = genes, columns = 1 gene-name col + {len(tumor_cols)} tumor samples)")
```

```python
# Quick sanity check — preview the filtered data
print("=== Filtered Expression Data (first 5 rows, first 5 columns) ===")
expression_tumor.iloc[:5, :5]
```

# Step 4: Reshape Expression Data — Samples as Rows
Right now each **column** is a sample and each **row** is a gene.
We transpose so each **row** becomes a sample and each **column** becomes a gene.
This is the standard layout for machine learning (one row per patient).

```python
# Set the gene name column as the index so gene names become column headers after transpose
# gene_col[0] is the name of the gene column we identified earlier (e.g. "gene_id" or "gene_name")

expression_tumor = expression_tumor.set_index(gene_col[0])

# Transpose: rows ↔ columns
# Before: rows=genes, columns=samples  →  After: rows=samples, columns=genes
expr_transposed = expression_tumor.T

print(f"Before transpose: {expression_tumor.shape}  (genes × samples)")
print(f"After transpose:  {expr_transposed.shape}  (samples × genes)")
print(f"\nFirst 3 rows × first 5 columns:")
expr_transposed.iloc[:3, :5]
```

```python
# Create a "sample_id" column from the index
# After transpose, the index holds the original column names (TCGA barcodes like TCGA-XX-XXXX-01)
expr_transposed = expr_transposed.reset_index()
expr_transposed = expr_transposed.rename(columns={"index": "sample_id"})

print("sample_id column — first 5 values:")
print(expr_transposed["sample_id"].head().tolist())
```

```python
# Extract patient_id from sample_id
# sample_id looks like: TCGA-XX-XXXX-01
# patient_id is the first 3 parts: TCGA-XX-XXXX (split on "-" and keep first 3)
# This lets us match expression data to clinical/follow-up data later

expr_transposed["patient_id"] = expr_transposed["sample_id"].apply(
    lambda x: "-".join(x.split("-")[:3])
)

print("sample_id vs patient_id — first 5 rows:")
print(expr_transposed[["sample_id", "patient_id"]].head().to_string(index=False))
```

```python
# Final shape summary
print("=== Reshaped Expression Data ===")
print(f"Shape: {expr_transposed.shape}")
print(f"  → {expr_transposed.shape[0]} samples (rows)")
print(f"  → {expr_transposed.shape[1]} columns (sample_id + patient_id + {expr_transposed.shape[1] - 2} genes)")
print(f"\nFirst 5 columns: {expr_transposed.columns[:5].tolist()}")
print(f"Last 5 columns:  {expr_transposed.columns[-5:].tolist()}")
```
